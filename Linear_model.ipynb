```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [2, 4, 6, 8, 10, 12]
y = [6, 10, 11, 14, 22, 25]
```

```python
x = np.array(x).reshape(-1,1)
x
```

```python
model = LinearRegression()
```

```python
model.fit(x,y)
```

```python
model.predict([[14]])
```

```python
model.score(x,y)
```
